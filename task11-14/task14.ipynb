```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

plt.style.use('ggplot')
```

```python
x = np.array([-1.11, -6.10, 2.42])
y = np.array([-2.29, -2.91])

sgm_x_sq = 2
sgm_y_sq = 1
alpha = 0.05

mx = np.mean(x)
my = np.mean(y)
n = len(x)
m = len(y)

se = np.sqrt(sgm_x_sq/n + sgm_y_sq/m)
delt = (mx - my) / se
crit = st.norm.ppf(1 - alpha)

print(f"Средние: x_bar = {mx:.3f}, y_bar = {my:.3f}")
print(f"Статистика критерия: delta = {delt:.3f}")
print(f"Критическая область G: delta >= {crit:.3f}")
print("Вывод: " + ("Отвергаем H0" if delt >= crit else "Нет оснований отвергнуть H0"))


quant = st.norm(0, se).ppf(1 - alpha)

X = np.linspace(-1, 5, 500)
Y = 1 - st.norm(X, se).cdf(quant)

plt.figure(figsize=(10, 5))
plt.plot(X, Y, color='#2980b9', lw=2)

plt.axhline(alpha, color='#c0392b', linestyle='--', label=f'Уровень значимости α = {alpha}')
plt.axvline(0, color='#27ae60', linestyle=':', label='H0 (a = b)')

plt.title("Функция мощности критерия W(δ)")
plt.xlabel("Истинная разность средних δ = a - b")
plt.ylabel("W(δ)")
plt.legend()
plt.tight_layout()
plt.show()
```
